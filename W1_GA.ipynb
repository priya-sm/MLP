##**Q7**
**Determine the number of samples (rows) in the dataset which has num_rooms = 3 and num_bathrooms = 3**

```python
sample1= df[(df['num_rooms'] == 3) & (df['num_bathrooms'] == 3)]
sample1.shape[0]
```

    In Python (and Pandas), the & operator is used for element-wise logical AND operations between two conditions, often when filtering a DataFrame.
    It is commonly used inside a Pandas operation to combine multiple conditions.
    Always wrap each condition in parentheses to avoid precedence issues and ensure the expression works as expected.

##**Q8**
**Determine the number of samples (rows) in the dataset which has num_rooms = 3 or num_bathrooms = 3**

```python
sample2= df[(df['num_rooms'] == 3) | (df['num_bathrooms'] == 3)]
sample2.shape[0]
```

    The | operator in Pandas is used for element-wise logical OR operations between two conditions.
    It allows you to filter rows in a DataFrame where at least one of the conditions is True.
    Always use parentheses around each condition to avoid precedence issues.







##**Q9**
**In the Year 2022, how many houses (rows) located in the Greenwich Locality have exactly num_room == 3, and facing towards either the North or East?**

```python
sample3 = df[(df['Year'] == 2022) &
              (df['Locality'] == 'Greenwich') &
              (df['num_rooms'] == 3) &
              ((df['Face'] == 'North') | (df['Face'] == 'East'))]
sample3.shape[0]
```

##**Q10**
**Calculate the number of samples(rows) that correspond to the month of August across all the years.?**

```python
sample4 = df[df['Date'].dt.month==8]
sample4.shape[0]
```

    dt.month: It is used to extract the month from a datetime column or series.
    This attribute is available when a column or Series has been converted to the datetime format.
        Purpose: Extracts the month (as an integer) from a datetime object.
        Usage: It returns an integer between 1 and 12, representing the month.
