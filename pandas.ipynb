>***as_frame=True***

#**Basic Data Inspection**


**1. head():**
- *Purpose:* Returns the first n rows of a DataFrame or Series.
- *Syntax:* DataFrame.head(n) or Series.head(n)
- *Parameters:*
n: The number of rows to return (default is 5).

```python
df.head()
```

```python
df.head(n=5)
```

```python
df.head(3)
```

**2. tail():**

- *Purpose:* Returns the last n rows of a DataFrame or Series.
- *Syntax:* DataFrame.tail(n) or Series.tail(n)
- *Parameters:*
n: The number of rows to return (default is 5).

```python
df.tail()
```

```python
df.tail(n=1)
```

```python
df.tail(2)
```

**3.shape**
- The shape attribute in Pandas is used to determine the dimensions of a DataFrame or Series. It returns a tuple containing the number of rows and columns in the object.

```python
df.shape
```

**4. columns Attribute:**

- The most direct way to get a list of column names is to use the columns attribute of the DataFrame:

```python
df.columns
```

**5. info()**
- *Shape*: The number of rows and columns in the DataFrame.
- *Dtypes*: The data types of each column.
- *Memory Usage*: The amount of memory used by the DataFrame.
- *Non-Null Count*: The number of non-null values in each column.

```python
df.info()
```

**5. Summary statistics:**

***.describe():***
- Provides a comprehensive summary of the numerical columns in a DataFrame.
- Includes statistics like count, mean, standard deviation, minimum, 25th percentile, 50th percentile (median), 75th percentile, and maximum.

**.mean():** Calculates the mean (average) of the values in a Series or DataFrame.

**.median():**Calculates the median (middle value) of the values in a Series or DataFrame.

**.mode():** Returns the mode (most frequent value) of the values in a Series or DataFrame.

**.min() and .max():** Calculate the minimum and maximum values in a Series or DataFrame.

**.std():** Calculates the standard deviation of the values in a Series or DataFrame.

**.var():** Calculates the variance of the values in a Series or DataFrame.

```python
df.describe()
```

```python
df.describe().T
```

```python
df.describe(percentiles=[0.2,0.6,0.8])
```

#**AND operators**

#**OR operation**
